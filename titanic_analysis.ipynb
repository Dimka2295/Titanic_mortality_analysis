7. Використовуючи цикл заміни кількість родичів, що перевищує число 5(п'ять) на значення "above 5"(понад п'ять).       
Запиши значення в новий стовпчик ʼrelatives_categoryʼ.     
Результат представ у вигляді таблиці, побудованої з використанням групування та агрегації:

```python
ds['relatives_category'] = ds['relatives'].astype(str)

for i in range(len(ds)):
    if ds.loc[i, 'relatives'] > 5:
        ds.loc[i, 'relatives_category'] = 'above 5'

ds.groupby('relatives_category')['relatives_category'].agg(total_category='count')
```

8. Необхідно вивести на екран статистику по відсотковому показнику пасажирів з кількістю родичів більше 5 відносно загального числа пасажирів для кожного з міст посадки. Для цього:
- порахуй загальну кількість пасажирів в кожному з міст посадки,
- знайди число пасажирів з кількістю родичів більше 5 в кожному з міст посадки,
- поділи ці два стовчики між собою, перетворивши результат у відсотки (ціле число).        
Відобрази статистику порахованих показників, згруповану по містах посадки.

```python
stat_city=ds.groupby('embark_town')['embark_town'].agg(total_city='count')
stat_city['above_5_relatives'] = ds[ds['relatives_category'] == 'above 5'].groupby('embark_town')['relatives_category'].agg('count')
stat_city['percentage'] = np.round(stat_city['above_5_relatives'] / stat_city['total_city'] * 100)
stat_city=stat_city.fillna(0)
stat_city
```

10. Створи новий стовпець, де вік представлено категорією, замість числа (наприклад: до 14 років, 14-34 роки, 35-59 років, 60 і більше років). Виконай задачу з використанням користувацької функції. Осіб з невідомим віком познач відповідно.

```python
def classify_age(age):
    if pd.isna(age):
        return 'unknown'
    elif age < 14:
        return 'child'
    elif age <= 34:
        return 'young'
    elif age <= 59:
        return 'adult'
    else:
        return 'old'

ds['age_group'] = ds['age'].apply(classify_age)
ds.groupby('age_group')['age_group'].agg(total_by_category='count')
```

11. Перевір, в якій віковій категорії була найвища смертність.     
Для цього рекомендується перетворити стовпець 'alive' в булевий тип.    
 Потім підрахувати загальну кількість пасажирів та кількість тих, хто не вижив.      
 Потім обчисли відносний показниках для кожної категорії.

```python
ds['not_alive']=ds['alive'] == 'no'
stat_alive=ds.groupby('age_group')['not_alive'].agg(total_group='count')
stat_alive['total_died'] = ds[ds['not_alive'] == True ].groupby('age_group')['not_alive'].agg('count')
stat_alive['death_percentage'] = np.round(stat_alive['total_died'] / stat_alive['total_group'] * 100)
stat_alive
```

*Бонусне завдання*         :          
Підготуй розгорнуту статистику смертності по категорії віку, класу квитка, рівню каюти та кількості родичів.      
Які фактори, на твою думку, найбільше пов'язані з рівнем смерності? (наприклад: найбільша смертність у відсотковому значенні спостерігається серед вікової групи ... класу квитка.... при наявності ... родичів та для рівня каюти.... Фактор ... має найвищий вплив на смертність)

```python
ds['deck'] = ds['deck'].astype(str).replace('nan', 'Unknown')
stat_alive_sorted=stat_alive.sort_values('death_percentage', ascending=False)

st2=ds.groupby('class', observed=False)['not_alive'].agg(class_group='count')
st2['died_by_class'] = ds[ds['not_alive'] == True ].groupby('class', observed=False)['not_alive'].agg('count')
st2['death_percentage2'] = np.round(st2['died_by_class'] / st2['class_group'] * 100)
st2_sorted=st2.sort_values('death_percentage2', ascending=False)

st3=ds.groupby('deck', observed=False)['not_alive'].agg(deck_group='count')
st3['died_by_deck'] = ds[ds['not_alive'] == True ].groupby('deck', observed=False)['not_alive'].agg('count')
st3['death_percentage3'] = np.round(st3['died_by_deck'] / st3['deck_group'] * 100)
st3_sorted=st3.sort_values('death_percentage3', ascending=False)

st4=ds.groupby('relatives_category', observed=False)['not_alive'].agg(rel_group='count')
st4['died_by_rel'] = ds[ds['not_alive'] == True ].groupby('relatives_category', observed=False)['not_alive'].agg('count')
st4['death_percentage4'] = np.round(st4['died_by_rel'] / st4['rel_group'] * 100)
st4_sorted=st4.sort_values('death_percentage4', ascending=False)

print('Stat alive by age:\n', stat_alive_sorted, '\n')
print('Stat alive by class:\n', st2_sorted, '\n')
print('Stat alive by deck:\n', st3_sorted, '\n')
print('Stat alive by relatives_category:\n', st4_sorted, '\n')
```

Найбільша смертність у відсотковому значенні спостерігається серед вікової групи 'old' класу квитка 'third' при наявності 5 родичів та для рівня каюти 'A' (якщо опустити невідомі каюти). Фактор великої кількості родичів має найвищий вплив на смертність у відсотках (можливо тому що затонула одразу вся сім'я), але якщо брати кількісно - то найвища смертність у старих та дітей(категорії 'old' та 'young', а також пасажирів 3-го класу (найдешевшого).
